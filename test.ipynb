#### Filling values
If column is object, we will fill it some string like 'unknown'. 
If column is number, we will fill it mean value 

```python
def fill_nulls(df, columns):
    for column in columns:
        column_type = df[column].dtype 
        if column_type == 'object':
            df[column].fillna('unknown', inplace=True)
        else:
            df[column].fillna(df[column].median(), inplace=True)
    return df
```

```python
columns = nullValuesPercent.index[4:]
columns
```

```python
all_data = fill_nulls(all_data,columns)
all_data.isnull().sum().sum()
```

too big matrix to analyze. Make analytic function

```python
def choose_columns_with_high_corr(corr_matrix, r):
    high_corr_cols = {}
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > r:
                colnames = tuple([corr_matrix.columns[i], corr_matrix.columns[j]])
                high_corr_cols[colnames] = corr_matrix.iloc[i, j]
    return high_corr_cols
```

```python
high_corr_cols = choose_columns_with_high_corr(corr_matrix, 0.99)
high_corr_cols = dict(sorted(high_corr_cols.items(), key=lambda item: item[1]))
high_corr_cols
```
